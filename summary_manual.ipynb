```python
from functions import *
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = create_league_dict(league_id)
```

```python
max_points_in_gw = 0
manager_pts = ''
for i in data:
    m = max(data[i]['points'])
    if m > max_points_in_gw:
        max_points_in_gw = m
        manager_pts = i

print(f'Most points in a game week:\n{manager_pts}: {max_points_in_gw}')
```

```python
most_hits = 0
manager_hits = ''
num_hits = 0
for i in data:
    m = sum(data[i]['hits'])
    if m > most_hits:
        most_hits = m
        manager_hits = i
        num_hits = int(m/4)

print(f'Most hits taken:\n{manager_hits}: {num_hits}')
```

```python
least_hits = 0
manager_hits = ''
num_hits = 100000
for i in data:
    m = sum(data[i]['hits'])
    if m < most_hits:
        least_hits = m
        manager_hits = i
        num_hits = int(m/4)

print(f'Least hits taken:\n{manager_hits}: {num_hits}')
```

```python
highest_gw_rank = 100000000000
manager_gw = ''

for i in data:
    m = min(data[i]['gw_rank'])
    if m < highest_gw_rank:
        highest_gw_rank = m
        manager_gw = i

print(f'Highest rank for a single game week:\n{manager_gw}: {highest_gw_rank}')
```

```python
highest_ov_rank = 100000000000
manager_ov = ''

for i in data:
    m = min(data[i]['overall_rank'])
    if m < highest_ov_rank:
        highest_ov_rank = m
        manager_ov = i
print(f'Highest overal rank after any game week:\n{manager_ov}: {highest_ov_rank}')
```

```python
most_bench_pts = 0
manager_bench = ''

for i in data:
    m = sum(data[i]['points_on_bench'])
    if m > most_bench_pts:
        most_bench_pts = m
        manager_bench = i
print(f'Most points on bench:\n{manager_bench}: {most_bench_pts}')
```

```python
least_bench_pts = 1000000
manager_bench_l = ''

for i in data:
    m = sum(data[i]['points_on_bench'])
    if m < least_bench_pts:
        least_bench_pts = m
        manager_bench_l = i
print(f'Least points on bench:\n{manager_bench_l}: {least_bench_pts}')
```

```python
most_consistent = 10000000000
most_consistent_manager = ''
avg_pts_m = 0

for i in data:
    m = np.std(data[i]['points'])
    if m < most_consistent:
        most_consistent = m
        most_consistent_manager = i
        avg_pts_m = np.mean(data[i]['points'])
print(f'Most consistent manager: {most_consistent_manager}\nAverage Points: {avg_pts_m}\nStandard Deviation: {np.round(most_consistent,2)}')
```

```python
least_consistent = 0
least_consistent_manager = ''
avg_pts_l = 0

for i in data:
    m = np.std(data[i]['points'])
    if m > most_consistent:
        least_consistent = m
        least_consistent_manager = i
        avg_pts_l = np.mean(data[i]['points'])

print(f'Least consistent manager: {least_consistent_manager}\nAverage Points: {avg_pts_l}\nStandard Deviation: {np.round(least_consistent,2)}')
```

```python

plt.figure(figsize=(16,10))
for i in data:
    plt.plot(data[i]['cumulative_points'], label = i, linewidth = 3)

plt.title('League Points', size = 16)
plt.xlabel('Game Week', size = 16)
plt.ylabel('Total Points', size = 16)
plt.legend(fontsize=16)
plt.savefig('points_worm.png')
```
