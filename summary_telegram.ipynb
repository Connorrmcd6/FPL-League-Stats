```python
from functions import *
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = create_league_dict(league_id)
```

```python
max_points_in_gw = 0
manager_pts = ''
for i in data:
    m = max(data[i]['points'])
    if m > max_points_in_gw:
        max_points_in_gw = m
        manager_pts = i
```

```python
most_hits = 0
manager_hits = ''
num_hits = 0
for i in data:
    m = sum(data[i]['hits'])
    if m > most_hits:
        most_hits = m
        manager_hits = i
        num_hits = int(m/4)
```

```python
highest_gw_rank = 100000000000
manager_gw = ''

for i in data:
    m = min(data[i]['gw_rank'])
    if m < highest_gw_rank:
        highest_gw_rank = m
        manager_gw = i
```

```python
highest_ov_rank = 100000000000
manager_ov = ''

for i in data:
    m = min(data[i]['overall_rank'])
    if m < highest_ov_rank:
        highest_ov_rank = m
        manager_ov = i
```

```python
most_bench_pts = 0
manager_bench = ''

for i in data:
    m = sum(data[i]['points_on_bench'])
    if m > most_bench_pts:
        most_bench_pts = m
        manager_bench = i
```

```python
most_consistent = 10000000000
most_consistent_manager = ''
avg_pts_m = 0

for i in data:
    m = np.std(data[i]['points'])
    if m < most_consistent:
        most_consistent = m
        most_consistent_manager = i
        avg_pts_m = np.mean(data[i]['points'])
```

```python
least_consistent = 0
least_consistent_manager = ''
avg_pts_l = 0

for i in data:
    m = np.std(data[i]['points'])
    if m > most_consistent:
        least_consistent = m
        least_consistent_manager = i
        avg_pts_l = np.mean(data[i]['points'])
```

```python

plt.figure(figsize=(16,10))
for i in data:
    plt.plot(data[i]['cumulative_points'], label = i)

plt.title('League Points', size = 16)
plt.xlabel('Game Week', size = 16)
plt.ylabel('Total Points', size = 16)
plt.legend(fontsize=16)
plt.savefig('points_worm.png')
```
